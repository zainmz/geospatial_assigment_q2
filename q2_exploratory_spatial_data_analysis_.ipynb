```python
import geopandas as gpd
import pandas as pd

# Load the hotel data
hotels_df = pd.read_excel('Classified Hotels( 1-5 Star).xlsx')
hotels_df['Grade'] = hotels_df['Grade'].map({'ONE': 1, 'TWO': 2, 'THREE': 3, 'FOUR': 4, 'FIVE': 5})

# Load the shapefile
shapefile_gdf = gpd.read_file('Shape File\lka_admbnda_adm1_slsd_20220816.shp')
```

```python
# Convert the hotels DataFrame to a GeoDataFrame
hotels_gdf = gpd.GeoDataFrame(
    hotels_df,
    geometry=gpd.points_from_xy(hotels_df.Logitiute, hotels_df.Latitude)
)

# Set the CRS for the hotels GeoDataFrame to match the shapefile
hotels_gdf.crs = shapefile_gdf.crs
```

```python
import matplotlib.pyplot as plt

# Plot the shapefile
fig, ax = plt.subplots(figsize=(10, 15))
shapefile_gdf.plot(ax=ax, color='white', edgecolor='black')

# Plot the hotels on top of the shapefile
hotels_gdf.plot(ax=plt.gca(), color='green', markersize=20)


plt.show()
```

```python
import folium
from folium.plugins import HeatMap

# Create a map centered around Sri Lanka
sri_lanka_heatmap = folium.Map(location=[7.8731, 80.7718], zoom_start=8)

# Extract the coordinates of the hotels for the heatmap
hotel_coords = hotels_gdf[['geometry']].copy()
hotel_coords['geometry'] = hotel_coords['geometry'].apply(lambda x: [x.y, x.x])
hotel_coords = hotel_coords['geometry'].tolist()

# Create a Heatmap layer and add it to the base map
HeatMap(hotel_coords).add_to(sri_lanka_heatmap)

sri_lanka_heatmap
```

```python
from libpysal.weights import KNN

# Calculate k-nearest neighbors weights
weights = KNN.from_dataframe(hotels_gdf, k=8)
weights.transform = 'R'
```

```python
from esda.moran import Moran
from esda.moran import Moran_Local

# Calculate global Moran's I
moran_global = Moran(hotels_gdf['Grade'], weights)

# Calculate local Moran's I (LISA)
moran_local = Moran_Local(hotels_gdf['Grade'], weights)
```

```python
import matplotlib.pyplot as plt
from libpysal.weights import lag_spatial
import seaborn as sns  # for nicer plots

# Calculate the spatial lag of the variable of interest
y = hotels_gdf['Grade']
w = weights
y_lag = lag_spatial(w, y)

# Calculate Moran's I
moran = Moran(y, w)
moran_i = moran.I
moran_p = moran.p_sim

# Create a Moran scatterplot
sns.regplot(x=y, y=y_lag, ci=None)
plt.axhline(y_lag.mean(), linestyle='--', color='grey')
plt.axvline(y.mean(), linestyle='--', color='grey')
plt.title(f"Moran Scatterplot (Moran's I: {moran_i:.2f}, p-value: {moran_p:.2f})")
plt.ylabel('Spatial Lag of Variable')
plt.xlabel('Grade')
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from esda.moran import Moran_Local
from splot.esda import plot_moran
from libpysal.weights import Queen


# Calculate LISA
local_moran = Moran_Local(hotels_gdf['Grade'].values, weights)

# Create a new column in hotels_gdf for the LISA cluster indicators
hotels_gdf['moran_loc'] = local_moran.q

# Plot the shapefile as the base layer
fig, ax = plt.subplots(figsize=(10, 10))
shapefile_gdf.plot(ax=ax, color='white', edgecolor='black')

# Plot significant clusters
sig = local_moran.p_sim < 0.05
hot_spots = sig * (local_moran.q == 1)  # High-high
cold_spots = sig * (local_moran.q == 3)  # Low-low

# Plot hot spots
hotels_gdf[hot_spots].plot(ax=ax, marker='o', color='red', markersize=15)

# Plot cold spots
hotels_gdf[cold_spots].plot(ax=ax, marker='o', color='blue', markersize=15)

# Set title and axes
ax.set_title('LISA Cluster Map')
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_aspect('equal')

# Remove axis clutter
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

plt.show()
```
